# Analisis Tingkat Churn Player Game "Gacha"
Tujuan : menganalisis kemungkinan player akan "churn".


## 1. Overview

```python
import pandas as pd

df = pd.read_csv("player-dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 2. Distribusi

```python
df["churn"].value_counts()
```

Distribusi menunjukkan jumlah player yang churn lebih banyak dibandingkan yang bertahan. Hal ini mengindikasikan bahwa dalam simulasi ini, kondisi yang memicu churn cukup dominan, sehingga dataset memiliki kecenderungan churn yang tinggi.

## 3. Analisis

### 3.1 Session Time

```python
df.groupby("churn")["session_time"].mean()
```

Player yang churn memiliki rata-rata session time yang lebih rendah dibandingkan player yang tidak churn. Hal ini menunjukkan bahwa tingkat engagement (waktu bermain) merupakan indikator kuat terhadap kemungkinan churn. Penurunan aktivitas bermain dapat menjadi sinyal awal bahwa player mulai kehilangan minat.

### 3.2 FPS / Hardware

```python
df.groupby("churn")["avg_fps"].mean()
```

Player yang churn cenderung memiliki average FPS yang lebih rendah. Hal ini menunjukkan bahwa performa device dan kenyamanan bermain memiliki pengaruh terhadap retensi player. Pengalaman bermain yang kurang optimal dapat menurunkan durasi bermain, yang pada akhirnya meningkatkan kemungkinan churn.

### 3.3 Pulls and SSR  Obtained

```python
df.groupby("churn")[["total_pulls", "ssr_obtained", "got_target"]].mean()
```

Player yang churn menunjukkan jumlah SSR yang relatif lebih tinggi dengan total pull yang lebih sedikit, namun memiliki tingkat keberhasilan mendapatkan karakter target yang lebih rendah. Hal ini mengindikasikan bahwa kepuasan pemain tidak hanya ditentukan oleh jumlah SSR yang diperoleh, tetapi oleh relevansi karakter tersebut terhadap preferensi pemain. Ketidakberhasilan mendapatkan karakter incaran dapat menimbulkan frustrasi, meskipun secara kuantitas hasil gacha terlihat baik.

### 3.4 Story Progress

```python
df.groupby("churn")["story_progress"].mean()
```

Player yang churn memiliki story progress yang sedikit lebih tinggi. Hal ini menunjukkan bahwa player yang sudah mencapai tahap lebih lanjut (mendekati endgame) cenderung mengalami penurunan minat, terutama jika tidak diimbangi dengan konten baru atau reward yang cukup. Kondisi ini dapat menyebabkan pemain merasa tidak memiliki tujuan lanjutan, sehingga meningkatkan kemungkinan churn.

### 3.5 Trend Session Time

```python
df.groupby("week")["session_time"].mean()
```

Rata-rata session time menunjukkan tren menurun seiring waktu dalam periode 8 minggu. Hal ini mengindikasikan adanya penurunan engagement secara bertahap, yang merupakan pola umum sebelum terjadinya churn. Penurunan aktivitas ini dapat menjadi indikator awal untuk mendeteksi player yang berpotensi berhenti bermain.

## 4. Insight and Conclusion

Berdasarkan hasil analisis, terdapat beberapa faktor utama yang mempengaruhi kemungkinan churn pada player game gacha:

- Engagement (Session Time) : 
  Waktu bermain merupakan indikator paling kuat. Player dengan durasi bermain rendah dan tren aktivitas yang menurun memiliki kemungkinan churn yang lebih tinggi
- User Experience (Hardware / FPS) :
  Performa device berpengaruh terhadap kenyamanan bermain. FPS yang rendah dapat menurunkan pengalaman bermain dan secara tidak langsung mengurangi durasi bermain, yang meningkatkan risiko churn. :
- Gacha Satisfaction (Target vs Quantity) :
  Kepuasan player tidak ditentukan oleh jumlah SSR yang diperoleh, tetapi oleh keberhasilan mendapatkan karakter target. Player yang tidak mendapatkan karakter yang diinginkan cenderung mengalami frustrasi dan lebih berpotensi churn.
- Progression (Story / Endgame) :
  Player yang sudah mencapai tahap lanjut (endgame) cenderung mengalami penurunan motivasi jika tidak tersedia konten atau reward yang cukup. Hal ini dapat menyebabkan churn meskipun progress permainan sudah tinggi.

Secara keseluruhan, churn dipengaruhi oleh kombinasi antara engagement, pengalaman bermain, kepuasan reward, dan progres permainan, bukan hanya satu faktor tunggal.

Analisis ini dilakukan pada data simulasi, sehingga hubungan antar variabel dibentuk berdasarkan asumsi tertentu. Namun, pendekatan ini mencerminkan bagaimana proses analisis data dilakukan dalam kasus nyata.
